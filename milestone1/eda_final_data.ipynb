### NaNs

```python
# Calculate number of NaN
def NaN_percentage(df):
    '''
    Returns a df with columnname and their respective number of NaN and what percentage this corresponds to.
    Args:
        df = movies, characters
    '''
    
    n_NaN = []
    perc_NaN = []
    length = len(df)

    for column in df:
        i_nan = df[column].isna().sum()
        perc_nan = i_nan/ length
        n_NaN.append(i_nan)
        perc_NaN.append(perc_nan)

    # Create df with results
    index_df = df.columns.to_list()
    df_NaN = pd.DataFrame({'number of NaN':n_NaN, 'percentage of NaN':perc_NaN}, index= index_df )
    return df_NaN
```

```python
data_nan_inner = NaN_percentage(data_all_inner)
data_nan_inner = data_nan_inner.sort_values(ascending=[True], by=['percentage of NaN'])

data_nan_outer = NaN_percentage(data_all_outer)
data_nan_outer = data_nan_outer.sort_values(ascending=[True], by=['percentage of NaN'])


f, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 10),sharey=True)

ax1.barh(data_nan_inner.index, data_nan_inner['percentage of NaN'])
ax1.set_title('Merge with inner join on countries')
ax1.title.set_size(20)
ax1.grid(axis='x')
ax1.tick_params(axis='x', labelsize=15)
ax1.tick_params(axis='y', labelsize=15)

ax2.barh(data_nan_outer.index, data_nan_outer['percentage of NaN'])
ax2.set_title('Merge with left join on countries')
ax2.grid(axis='x')
ax2.title.set_size(20)
ax2.tick_params(axis='x', labelsize=15)
ax2.tick_params(axis='y', labelsize=15)

plt.suptitle('Percentage of NaNs in the merged datasets by features', fontsize=20)
plt.savefig('./plots/Percentage of NaNs in the main dataset by features', bbox_inches = "tight")
```

### Correlation matrix


```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr = data_all_outer.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
ax.title.set_size(20)
ax.tick_params(axis='x', labelsize=15)
ax.tick_params(axis='y', labelsize=15)

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap = cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title('Correlation matrix of the datasets features', fontsize=20)
plt.savefig('./plots/correlation matrix', bbox_inches = "tight")
```
