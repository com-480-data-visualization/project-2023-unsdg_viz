#### Generator_expressions
Creating long lists can have large memory overhead. Often, we don't need to hold the entire lists in memory, but only need the elements one by one -- this is the case with all reductions, for example, such as the sum we used above.

"generator expressions" only creates the items when they are requested, computing them on the fly. -> use () instead of [] for list creation.

downside is that the elements of a generator expression can be accessed exactly once, i.e. there is no indexing!

```python
sum(x for x in range(1000000)) # instead: sum([x for x in range(1000000)])
sum(x for x in range(1000000) if x % 2 == 0)
```

#### P value
low p-value does not necessarily indicate that the findings are important or meaningful in a practical sense. A low p-value simply indicates that the data are unlikely to have occurred by chance alone if the null hypothesis is true. It is also possible that even if p-value is low but the effect size (difference of means) is small it might not be of practical significance.

Additionally, while a low p-value can provide evidence against the null hypothesis, it does not provide evidence in favor of the alternative hypothesis. To understand the strength of evidence in favor of the alternative hypothesis, one should look at effect size, or the magnitude of the difference or association, as well as confidence intervals that provide an interval estimate of the true value of the effect.

#### t-test
two-sided test: determine whether there is a significant difference in the means of two independent samples.

returns: Measure of the difference between the means of the two samples, and the two-tailed p-value.

t-statistic is a ratio of the difference between the means of the two samples and the standard deviation of the difference. A large t-statistic indicates that the means of the two samples are far apart, while a small t-statistic indicates that the means are similar.

p-value is used to determine whether the difference between the means is statistically significant with Null hypothesis that mean is same. A small p-value (typically less than 0.05) cannot reject the null hypothesis. Low p value indicates strong evidence against null hypothesis that there is no relationship between the two variable and is in favour that there is a correlation. 

Assumes that two samples have <b> equal variances and are normal distributed</b>, can be changed with variable 'equal_var'. 
When the variances of the two populations are not equal (i.e., the data is heteroscedastic), the test statistic of the t-test is no longer distributed normally, and the p-value calculated from the t-test may not be accurate. In this case, the t-test is known as an "unequal variances t-test" or "Welch's t-test" (This test is a conservative test, meaning it is less likely to reject the null hypothesis when it is false.)

alternative parameter is used to specify the type of alternative hypothesis to be tested.
```python
def confidence_interval(values, conf_lvl=0.95):
    return (np.quantile(values, q=(1 - conf_lvl)/2), np.quantile(values, q=(conf_lvl + 1)/2))
```

R^2:<br>
proportion of the variation in the dependent variable that is predictable from the independent variable(s). varies between 0 and 1, with larger numbers indicating better fits and 1 representing a perfect fit.<br>
**Advantages**: Gives an idea of the proportion of variation in the response variable that is explained by the predictor variables, easy to interpret<br>
**Disadvantages**:does not provide information about the direction of the error (i.e., whether the model is systematically underestimating or overestimating the response variable). Can be artificially high for models with many predictor variables, even if the model is not explaining much of the variation in the response variable. 

norm of residuals^2:<br>
calculated as the square-root of the sum of squares of residuals. The norm of residuals varies from 0 to infinity with smaller numbers indicating better fits and zero indicating a perfect fit.<br>
**Advantages**:scale-invariant measure, meaning that it is not affected by the units of measurement of the response or predictor variables,can be used to compare the performance of different models.<br>
**Disadvantages**:does not provide information about the direction of the error (i.e., whether the model is systematically underestimating or overestimating the response variable), 






More about R2: <br>
Imagine a scenario where a model with 100 predictor variables is fit to a dataset with only 10 observations. Even if only a few of the predictor variables are actually useful in explaining the response variable, the model will still likely have a high R2 value simply because it has so many variables to work with. This is an example of overfitting, where the model has a high degree of flexibility and can fit the noise in the data, resulting in high R2, but not generalize well to new data.

Another reason is when we have multicollinearity in our predictors, this means that the predictors are highly correlated with each other, in this case, R2 might be artificially high, even though the predictors might not have any real relationship with the outcome variable.

It's worth noting that R2 should not be used as the sole criterion for evaluating the performance of a model, it should be used in conjunction with other measures such as the norm of residuals and cross-validation, to get a more complete picture of the model's performance.

#### Helperfunctions-score

```python
def compute_confusion_matrix(true_label, prediction_proba, decision_threshold=0.5): 
    
    predict_label = (prediction_proba[:,1]>decision_threshold).astype(int)   
                                                                                                                       
    TP = np.sum(np.logical_and(predict_label==1, true_label==1))
    TN = np.sum(np.logical_and(predict_label==0, true_label==0))
    FP = np.sum(np.logical_and(predict_label==1, true_label==0))
    FN = np.sum(np.logical_and(predict_label==0, true_label==1))
    
    confusion_matrix = np.asarray([[TP, FP],
                                    [FN, TN]])
    return confusion_matrix
```

```python
def plot_confusion_matrix(confusion_matrix):
    [[TP, FP],[FN, TN]] = confusion_matrix
    label = np.asarray([['TP {}'.format(TP), 'FP {}'.format(FP)],
                        ['FN {}'.format(FN), 'TN {}'.format(TN)]])
    
    df_cm = pd.DataFrame(confusion_matrix, index=['Yes', 'No'], columns=['Positive', 'Negative']) 
    
    return sn.heatmap(df_cm, cmap='YlOrRd', annot=label, annot_kws={"size": 16}, cbar=False, fmt='')
```

```python
def compute_all_score(confusion_matrix, t=0.5):
    [[TP, FP],[FN, TN]] = confusion_matrix.astype(float)
    
    accuracy =  (TP+TN)/np.sum(confusion_matrix)
    
    precision_positive = TP/(TP+FP) if (TP+FP) !=0 else np.nan
    precision_negative = TN/(TN+FN) if (TN+FN) !=0 else np.nan
    
    recall_positive = TP/(TP+FN) if (TP+FN) !=0 else np.nan
    recall_negative = TN/(TN+FP) if (TN+FP) !=0 else np.nan

    F1_score_positive = 2 *(precision_positive*recall_positive)/(precision_positive+recall_positive) if (precision_positive+recall_positive) !=0 else np.nan
    F1_score_negative = 2 *(precision_negative*recall_negative)/(precision_negative+recall_negative) if (precision_negative+recall_negative) !=0 else np.nan

    return [t, accuracy, precision_positive, recall_positive, F1_score_positive, precision_negative, recall_negative, F1_score_negative]
```

#### Evaluation parameter
Parameter for binary classification: Precision P', 'Recall P', 'F1 score P', 'Precision N', 'Recall N', 'F1 score N'

**Accuracy**: This is the proportion of correctly classified instances out of the total number of instances. Advantage: it is easy to understand and interpret, it works well when the classes are balanced. When every class is equally important to you. Disadvantage: it can be **misleading when the classes are imbalanced**.

Precision: The number of instances that are relevant, out of the total instances the model retrieved.<br>
**Precision P**: The proportion of true positive predictions out of all the positive predictions made by the model. It is calculated as (TP) / (TP + FP). High precision means that the model makes a high proportion of correct positive predictions. likelihood that, if you have gotten a positive test result, you actually have the disease. Advantage: It is useful when the goal of the model is to minimize false positives. Disadvantage: It does not take into account the number of false negatives.<br>
**Precision N**: The proportion of true negative predictions out of all the negative predictions made by the model. It is calculated as (TN) / (TN + FN). High precision means that the model makes a high proportion of correct negative predictions.Advantage: It is useful when the goal of the model is to minimize false negatives. Disadvantage: It does not take into account the number of false positives.

**False Positive Rate**: FP / (FP+TN)<br>
**False negative rate**: FN / (FN+TP)

Recall: How many relevant items are retrieved?
**Recall P = true positive rate (TPR) = sensitivity**: The proportion of true positive predictions out of all the actual positive examples. It is calculated as (TP) / (TP + FN). High recall means that the model correctly identifies a high proportion of all the positive examples.<br>
**Recall N = True negative rate = specificity**: The proportion of true negative predictions out of all the actual negative examples. It is calculated as (TN) / (TN + FP). High recall means that the model correctly identifies a high proportion of all the negative examples.

**F1 score P**: The harmonic mean of precision and recall. It is calculated as 2 * (Precision * Recall) / (Precision + Recall). F1 score is a trade-off between precision and recall, and it is used to balance the two. High F1 score means that the precision and recall are balanced. Advantage: it is a balance between precision and recall and it is **useful when the classes are imbalanced**. Disadvantage: if precision and recall are both low, the F1-score will be low as well.<br>
**F1 score N**: The harmonic mean of precision and recall for N class. It is calculated as 2 * (Precision N * Recall N) / (Precision N + Recall N). F1 score is a trade-off between precision and recall, and it is used to balance the two for the N class. High F1 score means that the precision and recall are balanced for the N class.

**AUC-ROC Curve**: It is a graphical representation of the performance of a binary classification model, it plots the true positive rate against the false positive rate. ROC curve is not a good visual illustration for highly imbalanced data, because the False Positive Rate ( False Positives / Total Real Negatives ) does not drop drastically when the Total Real Negatives is huge. You should use it when you care equally about positive and negative classes.

**PR-AUC**:  the average of precision scores calculated for each recall threshold. Good when your data is heavily imbalanced. The intuition is the following: since PR AUC focuses mainly on the positive class (PPV and TPR) it cares less about the frequent negative class. When when you care more about positive than negative class. If you care more about the positive class and hence PPV and TPR you should go with Precision-Recall curve and PR AUC (average precision).

**Kappa statistic**: it is a measure of agreement between the predicted and true labels, it takes into account the agreement by chance. Advantage: it is **useful when the classes are imbalanced, it is independent of the class distribution**. Disadvantage: it is **sensitive to small changes** in the classification, it does not provide information about which class is being misclassified.

### Propensity score matching

Estimates the causal effect of a treatment on an outcome in observational studies. It is often used when a randomized controlled trial (RCT) is not possible or ethical.

Idea: match each individual in the treatment group to one or more individuals in the control group based on their "propensity score". The propensity score is the probability that an individual will receive the treatment, given their observed characteristics. By matching individuals with similar propensity scores, PSM aims to create a control group that is similar to the treatment group in terms of observed characteristics, and thus reduce the influence of confounding factors (covariates) on the estimate of the treatment effect.


Limitations: assumption that the propensity score correctly balances the observed characteristics between the treatment and control group, and that the model used to estimate the propensity score is correctly specified.

One way to check if a feature is balanced based on the propensity score is to compare the distribution of the feature in the treatment and control groups before and after matching. If the distribution is similar before and after matching, then the feature is considered to be balanced. (The groups are differentiated with feature x-class (E.g. treated, not treated))

Propensity score: represents its probability of x-class = value1, based on other features.

#### Get propensity score

```python
# let's standardize the continuous features
df_x['feature1'] = (df_x['feature1'] - df_x['feature1'].mean())/df_x['feature1'].std()


mod = smf.logit(formula='y ~  x1 + x2 + C(x3) + C(x4)  + C(x5)', data=df_x) # C for categorical features

res = mod.fit()

# Extract the estimated propensity scores
df_x['Propensity_score'] = res.predict()
```

#### Balancing the dataset via matching
Also see [max matching](#max-matching) <br>
Match point from the treated group with exactly most similar data-point from the control group (each  point from the control group is matched with max 1 point from the treated group). The sum (over all matched pairs) of absolute propensity-score differences between the two matched subjects should be minimized.

A maximum weight matching is a matching with the largest possible weight.

```python
# Function for calculating similarity of both scores
def get_similarity(propensity_score1, propensity_score2):
    '''Calculate similarity for instances with given propensity scores'''
    return 1-np.abs(propensity_score1-propensity_score2)
```

```python
# Separate the treatment and control groups
treatment_df = df_x[df_x['treat'] == 1]
control_df = df_x[df_x['treat'] == 0]

# Create an empty undirected graph
G = nx.Graph()

# Loop through all the pairs of instances
for control_id, control_row in control_df.iterrows():
    for treatment_id, treatment_row in treatment_df.iterrows():

        # Calculate the similarity 
        similarity = get_similarity(control_row['Propensity_score'],
                                    treatment_row['Propensity_score'])

        # Add an edge between the two instances weighted by the similarity between them
        G.add_weighted_edges_from([(control_id, treatment_id, similarity)])

# Generate and return the maximum weight matching on the generated graph
matching = nx.max_weight_matching(G)
```

Clusters only have convex shapes

#### Build model
The silhouette score is a measure of how similar an object is to its own cluster compared to other clusters. It ranges from -1 to 1, where a higher score indicates that the object is more similar to its own cluster than other clusters. The silhouette score can be used as a way to evaluate the quality of clustering and to determine the appropriate number of clusters for a dataset. The idea behind the silhouette score is that objects that are assigned to the correct cluster should have a higher similarity to the other objects in their cluster than to the objects in other clusters. It's important to notice that silhouette score assumes that the clusters are spherical, this means that clusters should have similar densities and similar diameters. If clusters have uneven densities or different diameters it may not work properly.

WCSS (within-cluster sum of squares) / Elbow method
The method is based on the idea that as the number of clusters increases, the within-cluster sum of squares (WCSS) will decrease, but at some point, the decrease in WCSS will begin to level off. The point at which the decrease in WCSS begins to level off is known as the "elbow point" and the number of clusters at this point is considered the optimal number of clusters. This method relies on the visual interpretation of the plot and the selection of the elbow point is somewhat subjective. Thus, it is a good idea to run it with multiple initial seeds for K-means, to ensure that the results are stable. Doesn't guarantee that it is the best solution, it just indicates that the increase in WCSS becomes less as k increases.

The dimensions of a dense bag of words matrix and a sparse bag of words matrix are the same. Both matrices have the same number of rows, which represents the number of documents, and the same number of columns, which represents the number of unique words in the corpus. The only difference is that in a dense matrix, each cell contains a numerical value representing the frequency of the word in the document, while in a sparse matrix, each cell contains a binary value representing the presence or absence of the word in the document.

However, the size of the matrix in terms of memory usage is different. Since sparse matrix only stores non-zero elements, it will take much less memory than dense matrix which stores all the element even if they are zero.

```python
STOPWORDS = spacy.lang.en.stop_words.STOP_WORDS

processed_docs = list()

# nlp.pipe: process in a memory-efficient manner, by processing the documents in small batches. 
# applies the pipeline of spaCy's NLP model to each document in the iterable, which includes various NLP tasks such as tokenization, part-of-speech tagging, named entity recognition, etc.
# batch_size: the number of documents to process at once. Increasing the batch size will speed up processing, but requires more memory.
# n_process: number of threads to use for parallel processing.


# If one process should not be applied, add before applying nlp.pipe
# nlp.remove_pipe('parser') or nlp.remove_pipe('tagger')

for doc in nlp.pipe(chunks, n_process=5, batch_size=10):

    # Process document using Spacy NLP pipeline.
    ents = doc.ents  # Named entities
    
    # Lemmatize tokens
    # Keep only words (no numbers, no punctuation): token.is_alpha returns a Boolean value of True if the token is composed of only alphabetic characters 
    # remove punctuation and remove stopwords.
    doc = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop]

    # Remove common words from a stopword list and keep only words of length 3 or more. Remove them again because lemantization could lead to new stopword
    doc = [token for token in doc if token not in STOPWORDS and len(token) > 2]

    # Add named entities, but only if they are a compound of more than word.
    doc.extend([str(entity) for entity in ents if len(entity) > 1])

    processed_docs.append(doc)
docs = processed_docs
del processed_docs
```

#### TF-IDF
Term Frequency — Inverse Document Frequency: compute a score for each word to signify its importance in the document and corpus. <br>
TF-IDF = Term Frequency (TF) * Inverse Document Frequency (IDF) with term = word <br>
With t — term (word), d — document (set of words), N — count of corpus, corpus — the total document set <br>
tf(t,d) = count of t in d / number of words in d <br>
idf(t) = log(N/(df + 1)) with df(t) = occurrence of t in N documents

IDF value of a word states how unique (important) it is to any particular document.

```python
# is_strongly_connected(graph) but manually
def is_strongly_connected(G):
    # Start DFS from an arbitrary node
    start_node = list(G.nodes())[0]
    visited = set()
    stack = [start_node]

    # DFS
    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            stack.extend(nx.all_neighbors(G, node))

    # Check if all nodes are reachable
    if len(visited) == G.number_of_nodes():
        print("The graph is strongly connected.")
        return True
    else:
        print("The graph is not strongly connected.")
        return False
```

**Idea:** Edges with high betweeness centrality separate communities. (betweenness centrality f= number of these shortest paths that pass through the node)

The betweenness of all existing edges in the network is calculated first.
The edge(s) with the highest betweenness are removed.
The betweenness of all edges affected by the removal is recalculated.
Steps 2 and 3 are repeated until no edges remain.

```python
# Calculates number of nodes (of category x) that falls into community1 and community2.

'''
graph is encoded within dictionary, where key of dictionary = Node-name. 
Dictionary has two values, category x and category y

Calculates number of nodes (of category x) that falls into community1 and community2.

returns: % of all nodes that belong to category x, % of all nodes that belong to category y, 
        number of nodes that belong to category x, number of nodes that belong to category y

Input:
    bisection: graph which is split into two by kernighan_lin_bisection function
                = tuple where tuple[0] = nodes in category x
    
    category: category x or y
                = one of the two categories from channel_cats (= value of dictionary)

    channel_cats: dictionary with key = that corresponds to a Node-name
'''
def calculate_community_probabilities(bisection, category, channel_cats):
    total_cat = len([x for x in channel_cats if channel_cats[x] == category])
    p_com1_cat = len([x for x in bisection[0] if channel_cats[x] == category]) / total_cat
    p_com2_cat = len([x for x in bisection[1] if channel_cats[x] == category]) / total_cat
    return p_com1_cat, p_com2_cat, p_com1_cat*total_cat, p_com2_cat*total_cat
```

#### Helper function
#### plot_degree_distribution, describe_graph, visualize graph

Degree distribution: probability distribution that describes the number of connections, or "degree," that each node in the graph has. The degree of a node is the number of edges that are connected to it.

The degree distribution is usually represented as a histogram, where the x-axis represents the degree of a node and the y-axis represents the probability of a node having that degree. The sum of all the probabilities in the distribution is equal to 1.

Different types of graphs have different degree distributions. For example, in a regular graph, all nodes have the same degree, so the degree distribution is a delta function (a spike at one value). In a random graph, the degree distribution is usually close to a normal distribution. In a scale-free network, a few nodes have a very high degree (called "hubs"), while most nodes have a low degree, resulting in a power-law distribution.

The degree distribution of a graph can be used to classify the graph into different types and to study the properties of the graph, such as its robustness and its connectivity.

```python
# Helper function for plotting the degree distribution of a Graph
def plot_degree_distribution(G):
    degrees = {}
    for node in G.nodes():
        degree = G.degree(node)
        if degree not in degrees:
            degrees[degree] = 0
        degrees[degree] += 1
    sorted_degree = sorted(degrees.items())
    deg = [k for (k,v) in sorted_degree]
    cnt = [v for (k,v) in sorted_degree]
    fig, ax = plt.subplots()
    plt.bar(deg, cnt, width=0.80, color='b')
    plt.title("Degree Distribution")
    plt.ylabel("Frequency")
    plt.xlabel("Degree")
    ax.set_xticks([d+0.05 for d in deg])
    ax.set_xticklabels(deg)
```

"Sparsity" of a graph with $n$ nodes is defined as follows: 

$ L = \frac{|E|}{|E_{max}|}$, where $E_{max} = \frac{n * (n-1)}{2}$

```python
# Helper function for printing various graph properties
def describe_graph(G):
    print(nx.info(G))
    if nx.is_connected(G):
        print("Avg. Shortest Path Length: %.4f" %nx.average_shortest_path_length(G))
        print("Diameter: %.4f" %nx.diameter(G)) # Longest shortest path
    else:
        print("Graph is not connected")
        print("Diameter and Avg shortest path length are not defined!")
    print("Sparsity: %.4f" %nx.density(G))  # #edges/#edges-complete-graph
    # #closed-triplets(3*#triangles)/#all-triplets
    print("Global clustering coefficient aka Transitivity: %.4f" %nx.transitivity(G))
```

```python
# Helper function for visualizing the graph
'''
    G: input graph which will be visualized
    with_labels: whether or not to draw labels on the nodes
    k: parameter controlling the distance between nodes in the layout. None -> a default value is used
    alpha: a value between 0 and 1 controlling the transparency of the edges. 
        A value of 1 means the edges are fully opaque, and a value of 0 means the edges are fully transparent.
    node_shape: the shape of the nodes in the graph. The default value is 'o' which means circular nodes.
        . would result in smaller circles
'''
def visualize_graph(G, with_labels=True, k=None, alpha=1.0, node_shape='o'):
    #nx.draw_spring(G, with_labels=with_labels, alpha = alpha)
    pos = nx.spring_layout(G, k=k)
    if with_labels:
        lab = nx.draw_networkx_labels(G, pos, labels=dict([(n, n) for n in G.nodes()]))
    ec = nx.draw_networkx_edges(G, pos, alpha=alpha)
    nc = nx.draw_networkx_nodes(G, pos, nodelist=G.nodes(), node_color='g', node_shape=node_shape)
    plt.axis('off')
```

#### Jaccard similarity

```python
def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))

# Note: similarity is computed 2 times! It can be optimized
similarity = []
for sr1 in spark_feature_50k:
    for sr2 in spark_feature_50k:
        similarity.append((sr1[0], sr2[0], jaccard_similarity(sr1[1], sr2[1])))


similarity_matrix_50k_words = pd.DataFrame(similarity).pivot(index=0, columns=1, values=2)
plot_heatmap(similarity_matrix_50k_words)
```

```python
from pyspark.ml.feature import RegexTokenizer, StopWordsRemover
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, col

# tokenize the text
regexTokenizer = RegexTokenizer(inputCol="feature1_body", outputCol="all_words", pattern="\\W")
spark_with_words = regexTokenizer.transform(spark_data1)

# remove stop words
remover = StopWordsRemover(inputCol="all_words", outputCol="words")
spark_with_tokens = remover.transform(spark_with_words).drop("all_words")

# get all words in a single dataframe
all_words = spark_with_tokens.select(explode("words").alias("word"))
# group by, sort and limit to 50k 
top50k = all_words.groupBy("word").agg(count("*").alias("total")).sort(col("total").desc()).limit(50000)

top50k.show()
```

```python
# Get all words per feature value

tokens = spark_with_tokens.select("feature", explode("words").alias("word")).distinct()
# Join with the whitelist of the top 50k
# results in DataFrame[feature: feature_value, word: string]
filtered_tokens = tokens.alias("t").join(top50k, tokens.word==top50k.word).select("t.*")

# Collect the representation of feature
spark_feature_50k = filtered_tokens.rdd.map(lambda r: (r.featurename, [r.word])).reduceByKey(lambda a,b: a+b).collect()
for sr in spark_feature_50k:
    print("featurename: {} - Words: {}".format(sr[0], len(sr[1])))
```
